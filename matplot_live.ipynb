```python
fig=plt.figure()
```

```python
ax1=fig.add_subplot(1,1,1)
```

```python
def animate(i):
    pullData=open('/home/vikas/sample.txt','r').read()
    dataArray=pullData.split('\n')
    xar=[]
    yar=[]
    for eachLine in dataArray:
        if len(eachLine)>1:
            x,y=eachLine.split(',')
            xar.append(int(x))
            yar.append(int(y))
   # axi.clear()
    ax1.plot(xar,yar)
ani=animation.FuncAnimation(fig,animate,interval=1000)
plt.show()
```
